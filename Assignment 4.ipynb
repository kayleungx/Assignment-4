```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup

url ="https://www.tripadvisor.com.hk/Attraction_Review-g294217-d2050166-Reviews-Peng_Chau_Island_Ping_Chau_Pingzhou-Hong_Kong.html"
qoute = requests.get(url)
soup = BeautifulSoup(qoute.content,'html.parser')

header = ['Spot','Type','Distance']
data = []
for item in soup.find_all("div", attrs={"class":"bteWY A Fi"}):
    spot = item.find ("div",attrs={"class":"WlYyy biNiR dRkMD dpKLb cQovs cWWWn ezpXT dDKKM"}).get_text()
    types = item.find ("span",attrs={"class":"eiJAX _a _b"})
    distance = item.find ("span",attrs={"class":"dmqNt"})
    print(spot)
    print(types)
    print(distance)
    data.append((spot,types,distance))

df = pd.DataFrame(data,
   columns = header) 
df.to_csv('pc_tripadvisor.csv', sep='\t', encoding='utf-8')
df
```
